```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import minimize
import sys
!{sys.executable} -m pip install nmrglue
import nmrglue
data = pd.read_table("normalized_all.dat", sep="\s+")
```

```python
#K alpha 1 & 2 deconvolution

# x0 - Center of the distribution / eV
x0_Ka1 = 2307.787
x0_Ka2 = 2306.515

# fwhm - Full-width at half-maximum of the Pseudo Voigt profile / eV
fwhm_Ka1 = 1.27
fwhm_Ka2 = 1.01

# tau - Lorentzian/Gaussian mixing parameter
tau_Ka1 = 0.5
tau_Ka2 = 0.2

# coefficients
beta_Ka1 = 0.7
beta_Ka2 = 0.26

def TSS(x):
    y = ((nmrglue.lineshapes1d.sim_pvoigt_fwhm(data['E'], x[0], x[1], x[2])*x[6]+nmrglue.lineshapes1d.sim_pvoigt_fwhm(data['E'], x[3], x[4], x[5])*x[7]-data['FeS2'])**2).sum()
    return(y)

res = minimize(TSS,[x0_Ka1,fwhm_Ka1,tau_Ka1,x0_Ka2,fwhm_Ka2,tau_Ka2,beta_Ka1,beta_Ka2])

print('Fit parameters')
print(res.x)

fig, ax = plt.subplots()
ax.plot(data['E'],data['FeS2'])
ax.plot(data['E'],nmrglue.lineshapes1d.sim_pvoigt_fwhm(data['E'], res.x[0], res.x[1], res.x[2])*res.x[6]+nmrglue.lineshapes1d.sim_pvoigt_fwhm(data['E'], res.x[3], res.x[4], res.x[5])*res.x[7],ls='--')
ax.plot(data['E'],nmrglue.lineshapes1d.sim_pvoigt_fwhm(data['E'], res.x[0], res.x[1], res.x[2])*res.x[6])
ax.plot(data['E'],nmrglue.lineshapes1d.sim_pvoigt_fwhm(data['E'], res.x[3], res.x[4], res.x[5])*res.x[7])
ax.legend(['Mixture','Fit','Ka1','Ka2'])
plt.xlabel('Energy, eV')
plt.ylabel('Intensity, a.u.')
plt.show()

#Unknown composition mixture fit
f_FeS2=0.8
f_Na2SO3=0.1
f_Na2SO4=0.1

def TSS(x):
    y = ((data['Mixture']-x[0]*data['FeS2']-x[1]*data['Na2SO3']-x[2]*data['Na2SO4'])**2).sum()
    return(y)

def con(t):
    return t[0] + t[1] - t[2]

res = minimize(TSS,[f_FeS2,f_Na2SO3,f_Na2SO4],constraints={'type':'eq', 'fun': con})

fig, ax = plt.subplots()
ax.plot(data['E'],data['Mixture'])
ax.plot(data['E'],data['FeS2']*res.x[0]+data['Na2SO3']*res.x[1]+data['Na2SO4']*res.x[2],ls='--')
ax.plot(data['E'],data['FeS2']*res.x[0])
ax.plot(data['E'],data['Na2SO3']*res.x[1])
ax.plot(data['E'],data['Na2SO4']*res.x[2])
ax.legend(['Mixture','Fit','FeS2','Na2SO3','Na2SO4'])
plt.xlabel('Energy, eV')
plt.ylabel('Intensity, a.u.')
plt.show()

print('Fraction of S-2: ', res.x[0])
print('Fraction of S+4: ', res.x[1])
print('Fraction of S+6: ', res.x[2])
```
